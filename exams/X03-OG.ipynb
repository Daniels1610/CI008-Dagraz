```python
import math
import string
from typing import Union

letters = string.ascii_uppercase + string.ascii_lowercase
```

```python
class GNode:
    id:str
    value:Union[int, float]
    leafs:list
    level:int 

    def __init__(self, id) -> None:
        self.id = id
        self.value = None
        self.nodes = []
        self.level = None
```

```python
class GameTree():
    order:int           # Tree's Order (Leaf nodes come in Pairs, Triples, Quartet setting)
    levels:int          # Tree's Height (MAX/MIN Levels)
    leafs:list          # Tree's Leaf Nodes
    lens:int            # Quantity of Leaf Nodes
    id_iterable:str     # Leaf Node's IDs
    tr:int              # Global leafs tracker
    game_tree:list      # Game Tree Structure
     
    def __init__(self, terminals:list) -> None:
        self.leafs = terminals
        self.lens = len(terminals)
        self.order = 0
        self.set_order()
        self.levels = int(math.log(self.lens, self.order))
        self.id_iterable = letters
        self.tr = 0
        self.game_tree = []


    def create(self):
        if (self.game_tree): self.game_tree = []
        for l in range(self.levels+1):
            nodes_level = []
            for _ in range(0, self.order ** l):
                if (l == self.levels):
                    for n in self.leafs:
                        nodes_level.append((self.id_iterable[self.tr], n)); self.tr += 1
                    break

                nodes_level.append(self.id_iterable[self.tr]); self.tr += 1

            self.game_tree.append(nodes_level)
        return self.game_tree

        
    def alfabeta(self, alpha=-math.inf, beta=math.inf):
        a = alpha; b = beta

        for i in range(0, self.lens, self.order):
            val = self.leafs[0 + i : self.order + i]
            for j in val:
                if (self.levels % 2 != 0): a = max(a , j) 
                else: b =  min(b, j)
                if (a >= b): break
            if (self.levels % 2 == 0):
                if b >= a: a = b; b = beta
            else:
                if a <= b: b = a; a = alpha
        return a


    # GETTERS
    @property
    def get_order(self):
        return self.order
    
    @property
    def get_levels(self):
        return self.levels
    
    @property
    def get_leafs(self):
        return self.leafs

    @property
    def get_lens(self):
        return self.lens
    
    @property
    def get_game_tree(self):
        return self.game_tree

    # SETTERS
    def set_order(self):
        for i in range(2, self.lens):
            if (self.lens % i == 0): self.order = i; break

g = GameTree([5,3,5,8,9,7,9,3,2,3,8,4,6,2,6,4,3,3,8,3,2,7,9,5,0,2,8])
g2 = GameTree([10,5,7,11,12,8,9,8,5,12,11,12,9,8,7,10])
g2.create()
print(g2.get_game_tree)
g2.alfabeta()
```
